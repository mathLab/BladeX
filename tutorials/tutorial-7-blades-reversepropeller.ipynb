## Tutorial 7: Retrieve blade's parameters from iges file using reversepropeller class

The goal of this tutorial is to show how to start from an iges file containing a blade to obtain its parameters, both for the sections (camber, thickness) and for the whole blade (chord lengths, pitch, rake, skew angles).

We start from some useful imports, as usual.

```python
import numpy as np
from bladex import Blade, NacaProfile, ReversePropellerBladeX, ReversePropeller
```

Then, we create the same blade as for Tutorial 6 and store it as iges file.

```python
# Create sections with NACAProfile
n_sections = 10
sections = np.asarray([NacaProfile(digits='4412', n_points=50) for i in range(n_sections)])

# Define blade parameters
radii = np.arange(1.0, 11.0, 1.0)
chord_lengths = np.array([0.05, 2.5, 3.5, 4, 4.5, 5, 5.5, 6, 6.5, 7])
pitch = np.arange(1., 11.)
rake = np.arange(0.1, 1.1, 0.1)
skew_angles = np.arange(1., 21, 2.)

# Create the Blade
blade = Blade(sections=sections,
        radii=radii,
        chord_lengths=chord_lengths,
        pitch=pitch,
        rake=rake,
        skew_angles=skew_angles)

# We build the blade to be able to store its iges version
# The method build() internally calls apply_transformations()
blade.build()

# Storing the blade in iges format
iges_filename='data/blade_reversepropeller.igs'
blade.export_iges(iges_filename)
```

Now that we have our starting iges file, we can retrieve the blades parameters. In BladeX you can either use the `ReversePropellerBladeX`, if the blade has been generated using BladeX, or `ReversePropeller` otherwise. The former is robust and more accurate for BladeX blades because it exploits known structure of the iges file.

```python
# We need to specify: 
#   i) the name of the iges file; 
#   ii) values of the radii in mm at which the parameters are to be extracted;
#   iii) how many points to use to reconstruct the sections for each radius value
# In this case we use the same radii values (extrema excluded) of the blade's generation for a better comparison
radii_reverse = np.arange(2.0, 10., 1.)
# This operation may take few minutes
reverse = ReversePropellerBladeX(iges_filename, 1000*radii_reverse, num_points_top_bottom=100)

# Now we can retrieve the blade's parameters
print(reverse.chord_length_list)
print(reverse.pitch_list)
print(reverse.rake_list)
print(reverse.skew_angles_list)

# Finally, we can store the parameters to a csv file
reverse.save_global_parameters('data/blade_parameters.csv')
```
