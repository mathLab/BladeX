# BladeX

## Tutorial 6: Generate and deform a blade modifying the parameters

The goal of this tutorial is to show how to generate a blade starting from its parameters, related both to the sections (camber, thickness) and to the whole blade (chord lengths, pitch, rake, skew angles).

We start from some useful imports, as usual.

```python
import numpy as np
from bladex import Blade, CustomProfile, NacaProfile
```

Then, we create the blade using `NacaProfile` sections for the sake of simplicity, and then we extract and deform the parameters from it. The blade is created and visualized also initializing all other global parameters.

```python
# Create sections with NACAProfile
n_sections = 10
sections = np.asarray([NacaProfile(digits='4412', n_points=50) for i in range(n_sections)])

# Define blade parameters
radii = np.arange(1.0, 11.0, 1.0)
chord_lengths = np.array([0.05, 2.5, 3.5, 4, 4.5, 5, 5.5, 6, 6.5, 7])
pitch = np.arange(1., 11.)
rake = np.arange(0.1, 1.1, 0.1)
skew_angles = np.arange(1., 21, 2.)

# Create the Blade
blade = Blade(sections=sections,
        radii=radii,
        chord_lengths=chord_lengths,
        pitch=pitch,
        rake=rake,
        skew_angles=skew_angles)

# Tranform coordinates from planar to cylindrical coordinates and reflect the blade
blade.apply_transformations(reflect=True)

# Plot the sections
blade.plot(elev=None, azim=None, outfile=None)
```

Now the deformation rates for different parameters are defined. Then, new deformed `CustomProfile` sections are generated using the deformed parameters.

```python
# Define rates of deformation for some section parameters
def_thickness = 1.2
def_camber = 1.1

# Generate a deformed blade with modified parameters
sections_deformed = []
for sec in sections:
    sec_deformed = CustomProfile(chord_perc=sec.chord_percentage,
            chord_len=sec.chord_length,
            thickness_max=sec.thickness_max*def_thickness,
            camber_max=sec.camber_max*def_camber,
            thickness_perc=sec.thickness_percentage,
            camber_perc=sec.camber_percentage)
    sections_deformed.append(sec_deformed)
```

Finally, a deformed blade is defined starting from the above defined sections and from other deformed global parameters.

```python
# Define rates of deformation for some blade parameters
def_pitch = 1.2
def_chord_len = 0.6

blade_deformed = Blade(sections=sections_deformed,
        radii=radii,
        chord_lengths=chord_lengths*def_chord_len,
        pitch=pitch*def_pitch,
        rake=rake,
        skew_angles=skew_angles)

# Tranform coordinates from planar to cylindrical coordinates and reflect the blade
blade_deformed.apply_transformations(reflect=True)

# Plot the sections
blade_deformed.plot(elev=None, azim=None, outfile=None)
```
